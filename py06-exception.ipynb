## 예외(exception)
- 코드를 실행하는 중에 발생한 오류이다.
- 오류가 발생하는 이유는 프로그램이 잘못 동작하는 것을 막기 위함이다.

## 예외처리
- 예외처리는 에러가 발생해되 실행을 중단하지 않고 계속 실행하게 해주는 방법이다.
- 예외처리 구조
```
 try:
    실행할 코드
 except:
    오류가 발생했을 때 처리하는 코드
 else:
    오류가 없을 때 처리하는 코드
 finally:
    무조건 처리하는 코드      
```
- 파이썬에서는 선언할 때 오류를 일일이 체크하지 않고, 사용할 때 에러가 발생하면 디버깅하는 스타일이다.
- EAFP
   - 허락보다는 용서를 구하기가 쉽다.(Easier to ask for forgiveness than permission)
   - 출처 : https://docs.python.org/3/glossary.html


- except 뒤에 에러 이름을 넣으면, 해당하는 에러가 발생했을 때만 except를 발생시킨다.
- Built-in Exceptions 예외 계층도 : https://docs.python.org/3/library/exceptions.html#exception-hierarchy 


````python
# try 블록이 정상적으로 실행되지 않으면 else 블록은 실행되지 않는다.
# try영역에서 예외가 발생했을 때: try 실행 -> except 실행 -> finally 실행
# try영역이 정상적으로 수행했을 때 : try 실행 -> else 실행 -> finally 실행

try:
    a = int('3')
    b = int('a')
    c = int('4')
except:
    print('숫자가 아닌 문자가 입력되었습니다.')
else:
    print(f'{a}와{b}와{c}의 합:{a+b+c}')
finally:
    print('program end')
````

## 다중 except
- 다중 except 선언이 가능하다.
- except뒤에 오는 as는 try 블록에서 발생한 이름을 받아온다.
```
  try:
     실행할 코드
  except 예외 as 변수:
     예외가 발생했을 때 처리하는 코드
  except 예외 as 변수:
     예외가 발생했을 때 처리하는 코드  
  except:
     예외가 발생했을 때 처리하는 코드  


````python
y = [10, 20, 30]

try:
    index, x = map(int, input('인덱스와 나눌 숫자를 입력하세요.').split())
    print(y[index] // x)
except ZeroDivisionError as e:
    print('숫자를 0으로 나눌 수 없습니다.', e)
except IndexError as e :
    print('잘못된 인덱스입니다.', e)
except Exception as e:
    print(e)
else:
    print('정상처리 되었습니다.')
finally:
    print('program end')
````

````python
try:
    x = int(input('3의 배수 숫자입력:'))
    
    if x % 3 != 0:
        raise Exception('3의 배수가 아닙니다.')
    
    print(x)
    
except Exception as e:
    print(e)
````

````python
class Bird:
    # @abstractmethod
    
    def fly(self):
        pass
        raise NotImplementedError

b = Bird()
b.fly()
````

````python
class Bird:
    def fly(self):
        raise NotImplementedError # 상속받은 자식 클래스에서 fly()을 오버라이딩 해줘야 한다.

class Eagle(Bird):
    def display(self):
        print('display')
        
    def fly(self): # flay() 오버라이딩을 안하면 오류 발생
        print('very fast')

eagle = Eagle()
eagle.fly()
````

````python

from abc import abstractmethod

class Bird:
  @abstractmethod
  def fly(self):
    pass


class Eagle(Bird):
  def display(self):
    print('display')


#   def fly(self):
#     print('very fast')


eagle = Eagle() # ABC을 상속받은 Bird 추상클래스이다. Bird을 상속받은 Eagle도 추상클래스이다.
eagle.fly()
````

## 사용자 예외 클래스 만들기

````python
class UserClass(Exception):
    def __str__(self):
        return '허용하지 않는 데이터입니다.'
````

````python
try:
    x = int(input('숫자입력:'))
    
    if x <= 0:
        raise UserClass
    
    print(x)
except Exception as e: # 구체적인 예외처리 (밑으로 갈수록 추상적인 예외처리)
    print(e)
except UserClass as e:
    print(e)
except ValueError as e: # 추상적인 예외처리
    print(e)
except BaseException as e:
    print(e)
````

````python
try:
    x = int(input('숫자입력:'))
    
    if x <= 0:
        raise UserClass
    
    print(x)

except UserClass as e:
    print(e)
except ValueError as e:
    print(e)
except Exception as e: # 다중 except을 사용할 때 부모클래스는 아래에 위치한다.
    print(e)
except BaseException as e:
    print(e)
````

````python
def test():
    try:
        print('try 실행1')
        return
        print('try 실행2')
    except Exception as e:
        print('except', e)
    else:
        print('else')
    finally:
        print('finally') # 중간에 return이 있더라도 finally는 무조건 실행 됨
````
