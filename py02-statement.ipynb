  연산자 (operator) : 어떠한 기능을 수행하는 기호(+, -, *, /, %, >, <, >=, <=, ==, != ...)
  피연산자(operand) : 연산자의 작업 대상 ( 변수, 상수, 리터널, 수식...)
 
  연산자의 종류
 1. 산술연산자 : +, -, *, /,//, %
 2. 비교연산자 : >, <, >=, <=, ==, !=
 3. 논리연산자 : and, or, not
 4. 대입연산자 : =, +=, -=, *=, /=,...
 5. 식별연산자 : is, is not



# 산술 연산자
- 산술 연산자는 수학적 연산을 수행
- +, -, *, /, %, **, //

````python
print(10 / 4) # 실수 나눗셈을 수행
print(10 // 4) # 정수 나눗셈을 수행

print(2 ** 4) # 2의 4제곱

# %는 정수나눗셈의 나머지결과를 얻는 연산자입니다.
print(4 % 3)
print(27 % 5)

print("-" * 40)

n1 = int(input("첫번째 정수: "))
n2 = int(input("두번째 정수: "))

print(n1, "과", n2, "의 나눗셈의 몫은", n1//n2,
      "이고, 나머지는", n1%n2, "입니다.")
````

# 식별 연산자 (Identity Operators)
- 식별 연산자는 두 객체가 같은 객체인지를 확인한다.
- is : 두 객체가 같은 객체인지 확인
- is not 두 객체가 다른 객체인지 확인
- 결과 : True, False

````python
a = [1, 2, 3]
b = a
c = [1, 2, 3]
print(a is b)  # True (a와 b는 같은 객체)
print(a is c)  # False (a와 c는 같은 값을 가지지만 다른 객체)
````

#   멤버 연산자 (Membership Operators)
- 멤버 연산자는 시퀀스 내에 값이 존재하는지 확인한다.
- in : 특정값이 멤버의 값인지 확인
- in not : 특정값이 멤버의 값인지 아닌지 확인
- 결과 : True, False


````python
a = [1, 2, 3]
print(2 in a)  # True
print(4 not in a)  # True # is not 과 차이 주의!
````

# 제어문(statement controller)
- 문장의 흐름을 제어해 준다.
- 조건문, 반복문
- 조건문
  - if~else
- 반복문
  - for ~ in
  - while
- break, continue


```python
if 조건식:
   수행할 문장
else:
   수행할 문장
```

````python
a = 8
if a > 5:
    print('OK')
else:
    print('Fail')
````

````python
a = 5
if a % 2 == 0: # 연산자 우선순위 적용 되어서 산술 연산자 먼저 실행 그 후에 비교 연산자 실행
    print('EVEN')
else:
    print('ODD')
````

````python
# 나이가 20이상이면 성인, 미만이면 청소년으로 처리
# 성인 남성입니다, 성인 여성입니다, 청소년 남성입니다, 청소년 여성입니다.
# 남, 여를 잘못입렸했으면 '다시 입력해 주세요.[남, 여]
age = int(input('나이:'))
gender = input('성별[남,여]:')

if age >= 20:
    if gender == "남성":
        print("성인 남성입니다.")
    elif gender == "여성":
        print("성인 여성입니다.")
    else:
        print("다시 입력해 주세요.[남, 여]")
else:
    if gender == "남성":
        print("청소년 남성입니다.")
    elif gender == "여성":
        print("청소년 여성입니다.")
    else:
        print("다시 입력해 주세요.[남, 여]")
````
