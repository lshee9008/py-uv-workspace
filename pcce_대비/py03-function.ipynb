## 함수(function)
 - 함수: 특정 목적을 가진 코드의 집합이며 독립적으로 호출될 수 있는 기능이다.
 - 함수 구조
  ```
    def 함수이름(parameter_1, parameter_2) :
        수행할 문장
        return 결과값
  ```
 - 함수 호출
  ```
     함수이름(argument_1, argument_2 )
  ```
   parameter : 인자, argument : 인수


````python
# 함수를 정의할 때 수행할 문장을 구현하지 않으면 에러가 발생한다.
def proc():
    pass # pass는 구현을 안할 때 오류가 발생하지 않도록 사용한다.

````

````python
print(proc())
````

````python
# 괄호를 붙이지 않으면 함수식이다.
# 
print(proc)
````

````python
a = proc
print(type(a))
print(a())
print(a.__name__) # a에 저장된 함수명을 확인할때 사용

````

## callable
 - 이름 뒤에( )를 붙이는 것을 call(호출)이라고 한다.
 - 이름 뒤에( )를 붙일 수 있는 것을 callable이라고 한다.
 - 함수는 사용할 때 이름 뒤에 괄호를 붙인다.
 - python에서 callable인 것은 3가지 유형이 있다.
   - (1)function
   - (2)class(init가 정의되었을 경우)
   - (3)object(call이 정의되었을 경우)


````python
k = 1
k()
````

````python
# callalble인지 아닌지 체크
callable(k)
````

````python
callable(proc)
````

````python
# 파라미터와 리턴값이 있는 함수 정의
def max(x, y):
    if x >= y:
        res = x
    else:
        ress = y
    return res
````

````python
print(max(10, 5))
````

````python
# 파라미터와 리턴값이 있는 함수 정의
def max(x, y):
    if x >= y:
        return x
    else:
        return y
````

````python
print(max(10, 5))
````

````python
# 파라미터와 리턴값이 없는 함수 정의
def max():
    x = 10
    y = 5
    if x >= y:
        print(x)
    else:
        print(y)
````

````python
print(max())
````
