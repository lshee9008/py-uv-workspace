## 함수(function)
 - 함수: 특정 목적을 가진 코드의 집합이며 독립적으로 호출될 수 있는 기능이다.
 - 함수 구조
  ```
    def 함수이름(parameter_1, parameter_2) :
        수행할 문장
        return 결과값
  ```
 - 함수 호출
  ```
     함수이름(argument_1, argument_2 )
  ```
   parameter : 인자, argument : 인수


````python
# 함수를 정의할 때 수행할 문장을 구현하지 않으면 에러가 발생한다.
def proc():
    pass # pass는 구현을 안할 때 오류가 발생하지 않도록 사용한다.

````

````python
print(proc())
````

````python
# 괄호를 붙이지 않으면 함수식이다.
# 
print(proc)
````

````python
a = proc
print(type(a))
print(a())
print(a.__name__) # a에 저장된 함수명을 확인할때 사용

````

## callable
 - 이름 뒤에( )를 붙이는 것을 call(호출)이라고 한다.
 - 이름 뒤에( )를 붙일 수 있는 것을 callable이라고 한다.
 - 함수는 사용할 때 이름 뒤에 괄호를 붙인다.
 - python에서 callable인 것은 3가지 유형이 있다.
   - (1)function
   - (2)class(init가 정의되었을 경우)
   - (3)object(call이 정의되었을 경우)


````python
k = 1
k()
````

````python
# callalble인지 아닌지 체크
callable(k)
````

````python
# 파라미터와 리턴값이 있는 함수 정의
def max(x, y):
    if x >= y:
        res = x
    else:
        ress = y
    return res
````

````python
print(max(10, 5))
````

````python
# 파라미터와 리턴값이 있는 함수 정의
def max(x, y):
    if x >= y:
        return x
    else:
        return y
````

````python
print(max(10, 5))
````

````python
# 파라미터와 리턴값이 없는 함수 정의
def max():
    x = 10
    y = 5
    if x >= y:
        print(x)
    else:
        print(y)
````

````python
max()
````

````python
print(max())
````

1. positional
 - 함수 호출 : 선언할 때 파라미터의 순서와 호출할때 argument의 입력순서를 그대로 매칭한다.

````python
def proc(a, b):
    return a, b # 튜플로 기호가 생략되어 있다.
````

````python
print(proc(2, 4))
````

2. keyword
- keyword 방식을 이용하면 순서와 상관없이 parameter 이름에 따라 값이 대입된다.

````python
print(proc(b=4, a=2))
````

3. keyword & positional 혼합

````python
def proc(a, b, c):
    return a, b, c
````

````python
print(proc(2, c=5, b=3)) # 한번 키워드를 지정한 매개변수가 있다면 뒤어있는 매개변수들도 다 키워드로 정의해야 한다.
````

````python
# 함수를 호출할때 keyword로 호출하면 이후는 전부 keyword로 호출해야 한다.
print(proc(2, c=5, 3))
````

4. positional only
- positional only는 함수를 사용할 때 positional방식으로만 parameter를 쓸 수 있도록 강제하는 방법이다.
- / 의 앞에 오는 parameter들은 positional방식으로 입력할 수 있지만,  keyword방식으로 입력할 수 없다.
- positional only 함수를 정의할 수 있는 것은 python 3.8부터 지원한다. 이전 버전에서는 사전 정의된 positional only 함수는 있지만, 사용자가 만들 수는 없었다.


````python
def proc(a, b, c, /): # 기호 앞에 있는 것들은 positional, 기호 뒤에 있는 것들은 keyword
    return a, b, c
````

````python
print(proc(4, 5, 6))
````

````python
print(proc(4, 5))
````

````python
print(proc(4, c=6, b=5))
````

````python
def proc(a, b, /, c, d): # 기호 앞에 있는 것들은 positional, 기호 뒤에 있는 것들은 positional 혹은 keyword
    return a, b, c, d
````

````python
print(proc(4, 5, 6, 7))
````

5. keyword only
- keyword only방식은 keyword방식으로만 입력하도록 강제하는 방법이다.
- '*' 이후로는 keyword 방식만 가능하다.


````python
def proc(*, a, b, c, d): # * 이후로 무조건 키워드임
    return a, b, c, d
````

````python
print(proc(a=3, b=5, c=4, d=6))
````

````python
# 현재 proc() 함수는 keyword only로만 사용해야 한다.
print(proc(3, 4, d=6, c=5))
````

````python
def proc(a, b, *, c, d):
    return a, b, c, d
````

````python
# 현재 proc() 함수는 keyword only로만 사용해야 한다.
print(proc(3, 4, d=6, c=5))
````

6. 가변 positional
- 가변 positional은 parameter 이름에 '*'를 붙인다.
- positional 방식으로 입력되는 argument의 갯수에 상관없이 모두 받아온다.

````python
def proc(*a): # *a는 튜플로 처리 예정
    print(type(a))
    return a
````

````python
print(proc(1))
````

````python
print(proc(1, 2, 3))
````

````python
# TypeError: proc() got an unexpected keyword argument 'a'
print(proc(1, 2, a=3))
````

````python
# 일반매개변수 뒤에 가변매개변수를 사용하고 가변매개변수 한韓 번만 사용한다.
def proc(a, *b):
    print(type(a))
    print(type(b))
    return a, b
````

````python
print(proc(1, 2, 3, 4, 5))
````

7. 가변 keyword
- 가변 keyword는 parameter 이름에 '**'를 붙인다.
- keyword방식으로 입력되는 argument의 개수에 상관없이 모두 받아온다.

````python
def proc(**a):
    print(type(a))
    return a
````

````python
proc(x=2, y=3)
````

````python
proc(X=3, y=1, z=[1, 2, 3]) # [list: sequence, mutable, hetergeneous]
````
