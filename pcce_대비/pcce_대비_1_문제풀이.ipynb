## 3번. [코드 작성] 반복문

1부터 `n`까지의 정수 중 3의 배수의 합을 구해 출력하는 코드를 작성하세요.

**입력**: `n = 10`
**출력**: `18` (3+6+9)

```python
n = int(input())

# 여기에 코드 작성

```

---

````python
n = int(input())

sum = 0

for i in range(1, n+1):
    if i % 3 == 0:
        sum += i
````

## 4번. [빈칸 채우기] 리스트

리스트 `numbers`에서 최댓값과 최솟값의 차이를 출력하는 코드입니다. 빈칸을 채우세요.

```python
numbers = [4, 8, 15, 16, 23, 42]

biggest = ______(numbers)
smallest = ______(numbers)

print(biggest - smallest)
```

**출력**: `38`

---

````python
numbers = [4, 8, 15, 16, 23, 42]

biggest = max(numbers)
smallest = min(numbers)

print(biggest - smallest)
````

## 7번. [빈칸 채우기] 2차원 리스트

3x3 리스트에서 대각선 원소들의 합을 구하는 코드입니다. 빈칸을 채우세요.

```python
board = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

total = 0
for i in range(3):
    total += board[i][______]

print(total)
```

**출력**: `15` (1+5+9)

---

````python
board = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

total = 0
for i in range(3):
    total += board[i][i]

print(total)
````

## 8번. [코드 작성] 딕셔너리

학생 이름과 점수가 담긴 딕셔너리에서 점수가 80점 이상인 학생의 이름만 리스트로 출력하는 코드를 작성하세요.

```python
scores = {"민수": 85, "지은": 72, "하늘": 91, "서준": 68}

# 여기에 코드 작성

```

**출력**: `['민수', '하늘']`

---

````python
scores = {"민수": 85, "지은": 72, "하늘": 91, "서준": 68}

result = [ i for i, v in scores.items() if v >= 80 ]

print(result)
````

## 9번. [디버깅] 조건문 + 반복문

다음 코드는 1부터 20까지의 수 중 소수(prime number)를 모두 출력해야 하는데, 오류가 있어 일부 소수가 아닌 값도 출력됩니다. 오류를 찾아 고치세요. 두 줄 고쳐야 함

```python
for n in range(1, 21):
    is_prime = True
    for i in range(1, n):
        if n % i == 0:
            is_prime = False
    if is_prime:
        print(n, end=" ")
```

**기대 출력**: `2 3 5 7 11 13 17 19`

---

````python
for n in range(1, 21):
    is_prime = True
    for i in range(1, n):
        if n % i == 0:
            is_prime = False
    
    if is_prime:
        print(n, end=" ")
    
````
