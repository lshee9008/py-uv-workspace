[문제2]   
문자열을 입력받아 각 문자들로 구성된 리스트로 반환하는 make_list 함수를 정의하세요.  
make_list("abcd")  

[출력결과]

['a', 'b', 'c', 'd']


```python
def make_list(aaaa):
    return [ i for i in aaaa]


print(make_list("abcd"))
```

[문제3]   
숫자로 구성된 하나의 리스트를 입력받아, 짝수들을 추출하여 리스트로 반환하는 pickup_even 함수를 구현하세요.   

pickup_even([3, 4, 5, 6, 7, 8])

[출력결과]   
[4, 6, 8]

```python
def pickup_even(li):
    return [ i for i in li if i % 2 == 0]

print(pickup_even([3, 4, 5, 6, 7, 8]))
```

[문제4]  
콤마가 포함된 문자열 숫자를 입력받아 정수로 변환하는 convert_int 함수를 정의하세요.

convert_int("1,234,567")

[출력결과]   
1234567

```python
def convert_int(stss):
    return int(stss.replace(",", ""))

print(convert_int("1,234,567"))
```

[문제5]    
2 와 5의 배수가 아닌 갯수와 합계를 구하는 함수를 구현하세요.    

resultr(1, 3, 5, 8, 9, 11, 15, 19, 18, 20, 30, 33, 31)  

[출력결과]    
2 와 5의 배수가 아닌 갯수 : 7    
2 와 5의 배수가 아닌 합계 : 107   

==============================================  
result(5,2,7,9,10,8,6)     

[출력결과]    
2 와 5의 배수가 아닌 갯수 :  2
2 와 5의 배수가 아닌 합계 :  16



```python
def result(*nums):
    count = 0
    total = 0

    for n in nums:
        if n % 2 != 0 and n % 5 != 0:
            count += 1
            total += n

    print("2와 5의 배수가 아닌 갯수 :", count)
    print("2와 5의 배수가 아닌 합계 :", total)
    
print(result(1, 3, 5, 8, 9, 11, 15, 19, 18, 20, 30, 33, 31))
```

문제6~13) 아래 함수를 호출했을때 정상적으로 실행이 되도록 호출하는 함수를 구현하세요.

```python
# 문제6)
def calc(a, b, /, c, *, d):
    return a + b + c + d

print(calc(1, 2, 3, d=4))
```

```python
# 문제7)
def introduce(name, /, age, *, country):
    return f"{name}, {age} years old from {country}"
```

```python
print(introduce("이승희", 26, country="South Korea"))
```

```python
# 문제8)
def f(x, *args, y, **kwargs):
    return x, args, y, kwargs
```

```python
print(f(1, 2, 3, 4, y=5, a=10, b=20))
```

```python
# 문제9)
def display(*args, **kwargs):
    print("ARGS:", args)
    print("KWARGS:", kwargs)
```

```python
display(1, 2, 3, name="이승희", age=26)
```

```python
# 문제10)
def config(*, debug=False, verbose=False):
    return f"Debug={debug}, Verbose={verbose}"
```

```python
print(config())
print(config(debug=True))
print(config(debug=True, verbose=True))
```
