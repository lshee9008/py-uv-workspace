```
[문제1]
분자(a)와 분모(b)를 받아 나눗셈 연산을 하되, 분모가 0일 때 프로그램이 정지하지 않고 "0으로 나눌 수 없습니다."를 반환하는 proc_divide 함수를 구현하세요.

[출력결과]
5.0
0으로 나눌 수 없습니다.


```

````python
def proc_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "0으로 나눌 수 없습니다."
    
        
    
print(proc_divide(10, 0))
````

```
[출력결과]
파일이 존재하지 않습니다.

```

````python
def proc_safe_open(path):
    try:
        with open(path, "r", encoding="utf-8") as file:
            print(file.read())
    except:
        print("파일이 존재하지 않습니다.")
        
proc_safe_open("test.txt")
````

```
[문제5]
리스트 내의 문자열 데이터 숫자들을 모두 더하는 과정에서 문자가 섞여 연산 오류(TypeError)가 발생하면, 무시하고 숫자 형태의 데이터만 합산하여 결과를 반환하는 proc_sum_list 함수를 구현하세요.


[출력결과]
60

```




````python
def proc_sum_list(original_list):
    total = 0
    
    for i in original_list:
        try:
            total += int(i)
        except ValueError:
            pass
        
    return total
    
print(proc_sum_list(["10", "20", "abc", "30"]))
````

```
[문제8]
하나의 try 블록 안에서 두 가지 이상의 에러가 발생할 여지가 있을 때, ValueError와 ZeroDivisionError를 멀티 except 구문으로 연쇄 캐칭하여 각각 다른 에러 안내 메시지를 출력하는 proc_multi_exception 함수를 구현하세요.

[출력결과]
에러: 숫자로 변환할 수 없는 입력입니다.
에러: 0으로 나눌 수 없습니다.

```

````python
def proc_multi_exception(a, b):
    try:
        num = int(a)
        print(num / b)
    except ValueError:
        print("에러: 숫자로 변환할 수 없는 입력입니다.")
    except ZeroDivisionError:
        print("에러: 0으로 나눌 수 없습니다.")
        
proc_multi_exception("abc", 2)
proc_multi_exception("10", 0)
````

```
[문제9]
문자열 형식의 숫자를 받아 나눗셈 연산을 수행하되, 예외가 발생하지 않았을 때만 작동하는 else 블록을 결합하여 "연산에 정상 성공했습니다." 문구와 결과값을 출력하는 proc_success_check 함수를 구현하세요.

[출력결과]
연산에 정상 성공했습니다. 결과: 10.0

```

````python
def proc_success_check(a, b):
    try:
        result = int(a) / int(b)
    except ValueError:
        print("숫자로 변환할 수 없습니다.")
    except ZeroDivisionError:
        print("0으로 나눌 수 없습니다.")
    else:
        print("연산에 정상 성공했습니다. 결과:", result)
        
proc_success_check("100", "10")
````

```
[문제10]
try-except 구조로 파일 처리 기능을 수행하면서, 작업 도중 예외 에러가 뜨거나 안 뜨거나 상관없이 언제나 함수 탈출 직전에 실행되어 확실하게 시스템 자원을 반환(close)해 주는 finally 블록을 포함한 proc_final_close 함수를 구현하세요.

[출력결과]
예외 핸들링: 파일을 찾지 못해 안전 모드로 전환합니다.
finally 블록: 파일 스트림 자원 체크를 완수합니다.
```


````python
def proc_final_close():
    file = None

    try:
        file = open("test.txt", "r")
        print(file.read())
    except FileNotFoundError:
        print("예외 핸들링: 파일을 찾지 못해 안전 모드로 전환합니다.")
    finally:
        if file:
            file.close()
        print("finally 블록: 파일 스트림 자원 체크를 완수합니다.")
        
proc_final_close()
````

```
[문제15]
학생 이름과 점수로 구성된 이중 리스트 데이터를 입력받아 'score.csv' 파일로 저장하는 proc_save_csv 함수를 구현하세요.

[출력결과]
(로컬 디렉토리에 'score.csv' 파일이 생성되고 데이터가 행 단위로 기록됨)

```

````python
import csv

def proc_save_csv(data):
    with open("score.csv", "w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)
        writer.writerows(data)
        
score = [
    ["홍길동", 90],
    ["이순신", 85],
    ["강감찬", 95]
]

proc_save_csv(score)
````

```
[문제16]
앞서 생성한 'score.csv' 파일을 읽어서 각 학생의 이름과 점수를 리스트 형태로 콘솔에 한 줄씩 출력하는 proc_read_csv 함수를 구현하세요.

[출력결과]
['홍길동', '90']
['이순신', '85']
['유관순', '95']
```

````python
import csv

def proc_read_csv():
    with open("score.csv", "r", newline="", encoding="utf-8") as file:
        reader = csv.reader(file)

        for row in reader:
            print(row)
            
proc_read_csv()
````

```
[문제18]
복잡하게 얽힌 딕셔너리 데이터를 인자로 받아 들여쓰기(공백 4칸) 옵션을 적용하여 가독성이 높은 'config.json' 파일로 보관하는 proc_save_json 함수를 구현하세요.

[출력결과]
(로컬 디렉토리에 들여쓰기가 적용된 'config.json' 파일이 생성됨)
```

````python
import json

def proc_save_json(data):
    with open("config.json", "w", encoding="utf-8") as file:
        json.dump(data, file, ensure_ascii=False, indent=4)
        
config = {
    "name": "홍길동",
    "age": 20,
    "city": "서울"
}

proc_save_json(config)
````

```
[문제20]
텍스트가 아닌 이미지 등 이진 파일을 다룰 수 있도록 바이너리 모드를 적용해 원본 파일을 읽고 대상 경로에 복사본을 생성하는 proc_file_copy 함수를 구현하세요.

[출력결과]
(로컬 디렉토리에 'copy.bin' 바이너리 파일이 복제되어 생성됨)

```

````python
def proc_file_copy(src, dest):
    with open(src, "rb") as source_file:
        data = source_file.read()

    with open(dest, "wb") as target_file:
        target_file.write(data)
        
# proc_file_copy("original.bin", "copy.bin")
````
