- 이름 뒤에 괄호( )를 붙일 수 있는 것을 callable이라고 한다.
- python에서 callable인 것이 3가지 유형  
 1 function  
 2 class(__init__가 정의되었을 경우)

````python
class FourCal:
  def __init__(self, a, b):
    print('init')
    self.a = a
    self.b = b
    
  def add(self):
    return self.a + self.b

  def sub(self):
    return self.a - self.b

  def mul(self):
    return self.a * self.b

  def div(self):
    return self.a // self.b
````

````python
fc = FourCal(4, 2)

print(fc.add())
print(fc.sub())
print(fc.mul())
print(fc.div())
````

### classmethod
- 첫 번째 인자로 클래스(cls)를 받는다.
- 클래스를 이용해 무언가를 해야할때 사용한다.
- 클래스 변수에 접근하거나 다른 클래스 메소드를 호출할 수 있다.
- 일반적으로 팩토리 메서드 등에 자주 사용된다.

````python
class B:
    x = 3 # 클래스 변수
    @classmethod
    
    def aa(cls): # classmethod는 관례상 cls를 사용한다.
        print('python')
        
    def bb(self):
        print('instance')
````

````python
B.x
````

````python
B.aa() # 클래스.클래스메소드
````

````python
B.bb()
````

````python
ab = B()
B.bb(ab)
````

````python
class B:
    x = 1
    
    @classmethod
    def aa(cls, x):
        cls.x = x
        
````

````python
B.x
````

````python
b = B()
````

````python
b.x #b 인스턴스에 x가 없으므로 B클래스의 x를 가져온다.
````

````python
b.x = 5 # b 인스턴스에 x변수를 생성한다.
````
