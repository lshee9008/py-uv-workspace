모듈?
- 모듈은 서로 연관된 함수, 변수, 클래스들을 모아놓은 '단일 파이썬 파일 (.py)'
특징
    - 코드의 재사용성을 높이고 관리를 쉽게 하기 위해 파일을 분리한 가장 작은 단위
    - import 키워드를 통해 다른 파일에서 불러와 사용 가능

모듈의 종류
    - 표준 모듈 : 파이썬에 기본으로 설치된 모듈과 패키지, 내장 함수를 묶어서 파이썬 표준 라이브러리
    - 사용자 생성 모듈 : 사용자가 만드는 모듈
    - 서드파티 : 협력 업체나 개인이 만들어서 제공하는 모듈

```python
%%writefile module_mymath_01.py

mypi = 3.14

def add(a, b):
    return a+b

def area(r):
    return mypi * r * r
```

## 모듈 가져오기(import)  


- 다른 프로그램으로부터 데이터를  갖고 오는 것을 의미한다.  
   
- import 방법




| 형식 | 의미 | 예제 |
| :--- | :--- | :--- |
| `import 모듈` | 모듈 전체를 가져오는 형식 | `import math` |
| `from 모듈 import *` | 모듈 전체를 가져오는 형식 | `from math import *` |
| `from 모듈 import 변수 or 함수 or 클래스` | 모듈 내에서 필요한 것만 가져오는 형식 | `from math import sin, cos, pi` |
| `import 모듈 as 별칭` | 모듈을 다른 이름으로 가져오는 형식 | `import numpy as np` |
| `from 모듈 import 변수 or 함수 or 클래스 as 별칭` | 모듈 내의 특정 요소를 다른 이름으로 가져오는 형식 | `from re import sub as substitute` |
| `from 모듈 import 변수 or 함수 or 클래스 as 별칭1, 변수 or 함수 or 클래스 as 별칭2` | 모듈 내의 여러 요소를 각각 다른 이름으로 가져오는 형식 | `from re import sub as sub1, subn as sub2` |
| `from 모듈 import (변수 or 함수 or 클래스, ...)` | 하나의 모듈에서 여러 개의 이름을 가져올 때 괄호를 사용하는 형식 | `from Tkinter import (Tk, Frame, Button, Entry)` |


```python
import module_mymath_01

print(dir(module_mymath_01))

print(module_mymath_01.mypi)
print(module_mymath_01.area(5))
```

import 모듈명 방식:  모듈 이름을 접두사로 붙여 변수와 함수에 명확하게 접근하는 표준 방식


from 모듈명 import 함수명 방식: 모듈명 접두사 없이 함수나 변수 이름만 가지고 곧바로 호출하는 간결한 방식


```python
from module_mymath_01 import mypi, add, area

print(mypi)
print(add(10, 20))
print(area(5))
```

import 모듈명 as 별명 방식 : 모듈 이름이 너무 길거나 복잡할 때 축약형 별칭(Alias)을 부여하는  방식

```python
import module_mymath_01 as mm

print(dir(mm))

print(mm.mypi)
print(mm.area(5))
```
